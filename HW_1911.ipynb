```python
def fuuu(s):
    bukvi_dict = {'б':'b','в':'v', 'г':'g', 'д':'d', 'ж':'zh', 'з':'z', 'и':'i',
                  'п':'p', 'р':'r', 'н':'n'}

#     s = 'Привет, Татьяна!'
    s = s.lower()
    translit_s = ''
    for letter in s:
    #     print(letter) # П
        if letter in bukvi_dict.keys():
            translit_s += bukvi_dict[letter]
        else:
            translit_s += letter
#     print(translit_s)
    
    return translit_s
    
```

```python
texts = ['привет, татьяна','Хай Иван', 'Пока-пока Максим']
for text in texts:
    print(fuuu(text))
```
